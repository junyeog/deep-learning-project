```python
import os
import tensorflow as tf
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# 1. 압축 해제된 이미지 파일들이 있는 디렉토리 경로 설정
data_dir = 'utkface_dataset/UTKFace'

# 2. 이미지 파일 경로와 라벨 추출
image_paths = []
genders = []

for img_name in sorted(os.listdir(data_dir)):
    if img_name.endswith('.jpg'):
        parts = img_name.split('_')
        if len(parts) >= 2:
            try:
                # gender 라벨 추출 (0: male, 1: female)
                gender = int(parts[1])
                image_paths.append(os.path.join(data_dir, img_name))
                genders.append(gender)
            except ValueError:
                # 파일명 파싱 오류 시 건너뛰기
                continue

# 리스트를 NumPy 배열로 변환
image_paths = np.array(image_paths)
genders = np.array(genders)

print(f"총 이미지 파일 개수: {len(image_paths)}")
print(f"성별 라벨 분포: 남성(0): {np.sum(genders == 0)}, 여성(1): {np.sum(genders == 1)}")

# 3. 훈련/검증/테스트 세트 분할.
X_train_paths, X_test_paths, y_train_genders, y_test_genders = train_test_split(
    image_paths, genders, test_size=0.2, random_state=42, stratify=genders
)
X_train_paths, X_val_paths, y_train_genders, y_val_genders = train_test_split(
    X_train_paths, y_train_genders, test_size=0.25, random_state=42, stratify=y_train_genders
) # 0.8 * 0.25 = 0.2 (전체 데이터의 20%를 검증 세트로)

print(f"\n훈련 이미지 개수: {len(X_train_paths)}")
print(f"검증 이미지 개수: {len(X_val_paths)}")
print(f"테스트 이미지 개수: {len(X_test_paths)}")

# 4. tf.data.Dataset을 위한 전처리 함수 정의
IMG_HEIGHT = 128
IMG_WIDTH = 128
CHANNELS = 3 # RGB 이미지

def preprocess_image(image_path, gender_label):
    # 이미지 파일 읽기
    img = tf.io.read_file(image_path)
    # 이미지 디코딩 (JPEG 형식, 3채널 RGB)
    img = tf.image.decode_jpeg(img, channels=CHANNELS)
    # 이미지 크기 조정
    img = tf.image.resize(img, [IMG_HEIGHT, IMG_WIDTH])
    # 픽셀 값 정규화 (0-1 범위)
    img = img / 255.0
    return img, gender_label

# 5. tf.data.Dataset 생성
BATCH_SIZE = 32 # 배치 크기 설정 (메모리 상황에 따라 16, 8 등으로 조절 가능)
AUTOTUNE = tf.data.AUTOTUNE # 자동 튜닝 (성능 최적화)

# 훈련 데이터셋
train_ds = tf.data.Dataset.from_tensor_slices((X_train_paths, y_train_genders))
train_ds = train_ds.map(preprocess_image, num_parallel_calls=AUTOTUNE) \
                   .shuffle(buffer_size=1024) \
                   .batch(BATCH_SIZE) \
                   .prefetch(buffer_size=AUTOTUNE)

# 검증 데이터셋
val_ds = tf.data.Dataset.from_tensor_slices((X_val_paths, y_val_genders))
val_ds = val_ds.map(preprocess_image, num_parallel_calls=AUTOTUNE) \
               .batch(BATCH_SIZE) \
               .prefetch(buffer_size=AUTOTUNE)

# 테스트 데이터셋
test_ds = tf.data.Dataset.from_tensor_slices((X_test_paths, y_test_genders))
test_ds = test_ds.map(preprocess_image, num_parallel_calls=AUTOTUNE) \
                .batch(BATCH_SIZE) \
                .prefetch(buffer_size=AUTOTUNE)

# 6. 딥러닝 모델 구축 (CNN 아키텍처)
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras import regularizers

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(IMG_HEIGHT, IMG_WIDTH, CHANNELS)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(), # 2D 특징 맵을 1D 벡터로 평탄화
    Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    Dropout(0.5), # 과적합 방지를 위한 Dropout
    Dense(1, activation='sigmoid') # 이진 분류를 위한 출력 레이어 (1개 뉴런, sigmoid)
])

# 7. 모델 컴파일
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.summary()

# 8. 콜백 설정
checkpoint_filepath = 'best_utkface_cnn_model.keras'
model_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_filepath,
    save_weights_only=False, # 가중치뿐 아니라 모델 전체 저장
    monitor='val_accuracy',  # 검증 정확도를 기준으로 모니터링
    mode='max',              # 정확도는 높을수록 좋으므로 'max'
    save_best_only=True,     # 최고 성능 모델만 저장
    verbose=1                # 진행 상황 출력
)

early_stopping_callback = tf.keras.callbacks.EarlyStopping(
    monitor='val_accuracy',  # 검증 정확도 모니터링
    patience=10,             # 10 에포크 동안 개선 없으면 중단
    mode='max',              # 정확도는 높을수록 좋으므로 'max'
    restore_best_weights=True, # 최적의 가중치로 복원
    verbose=1
)

reduce_lr_callback = tf.keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss',      # 검증 손실 모니터링
    factor=0.5,              # 학습률을 0.5배로 감소
    patience=5,              # 5 에포크 동안 개선 없으면 감소
    min_lr=0.000001,         # 최소 학습률
    verbose=1
)

callbacks_list = [model_checkpoint_callback, early_stopping_callback, reduce_lr_callback]

# 9. 모델 학습

history = model.fit(
    train_ds,
    epochs=50, # 충분히 큰 에포크 수 (EarlyStopping이 알아서 조기 중단)
    validation_data=val_ds, # 검증 데이터셋
    callbacks=callbacks_list,
    verbose=1 # 학습 진행 상황 출력
)



test_loss, test_accuracy = model.evaluate(test_ds, verbose=0)
print(f"\n테스트 손실: {test_loss:.4f}")
print(f"테스트 정확도: {test_accuracy:.4f}")


plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='훈련 정확도')
plt.plot(history.history['val_accuracy'], label='검증 정확도')
plt.title('CNN 모델 정확도')
plt.xlabel('에포크')
plt.ylabel('정확도')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='훈련 손실')
plt.plot(history.history['val_loss'], label='검증 손실')
plt.title('CNN 모델 손실')
plt.xlabel('에포크')
plt.ylabel('손실')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import os
import tensorflow as tf
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt


data_dir = 'utkface_dataset/UTKFace'

# 2. 이미지 파일 경로와 라벨 추출
image_paths = []
genders = []

for img_name in sorted(os.listdir(data_dir)):
    if img_name.endswith('.jpg'):
        parts = img_name.split('_')
        if len(parts) >= 2:
            try:
                gender = int(parts[1]) # 0: male, 1: female
                image_paths.append(os.path.join(data_dir, img_name))
                genders.append(gender)
            except ValueError:
                continue

image_paths = np.array(image_paths)
genders = np.array(genders)

print(f"총 이미지 파일 개수: {len(image_paths)}")
print(f"성별 라벨 분포: 남성(0): {np.sum(genders == 0)}, 여성(1): {np.sum(genders == 1)}")

# 3. 훈련/검증/테스트 세트 분할
X_train_paths, X_test_paths, y_train_genders, y_test_genders = train_test_split(
    image_paths, genders, test_size=0.2, random_state=42, stratify=genders
)
X_train_paths, X_val_paths, y_train_genders, y_val_genders = train_test_split(
    X_train_paths, y_train_genders, test_size=0.25, random_state=42, stratify=y_train_genders
)

print(f"\n훈련 이미지 개수: {len(X_train_paths)}")
print(f"검증 이미지 개수: {len(X_val_paths)}")
print(f"테스트 이미지 개수: {len(X_test_paths)}")


IMG_HEIGHT = 128
IMG_WIDTH = 128
CHANNELS = 3

def preprocess_image(image_path, gender_label):
    img = tf.io.read_file(image_path)
    img = tf.image.decode_jpeg(img, channels=CHANNELS)
    img = tf.image.resize(img, [IMG_HEIGHT, IMG_WIDTH])
    img = img / 255.0
    return img, gender_label

# 5. 데이터 증강 레이어 정의
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal"), # 좌우 반전
    tf.keras.layers.RandomRotation(0.1),     # -10% ~ +10% 범위 회전
    tf.keras.layers.RandomZoom(0.1),         # -10% ~ +10% 범위 확대/축소
    tf.keras.layers.RandomContrast(0.1),     # 밝기 대비 조절 (선택 사항)

])

BATCH_SIZE = 32
AUTOTUNE = tf.data.AUTOTUNE

# 훈련 데이터셋 (데이터 증강 적용)
train_ds = tf.data.Dataset.from_tensor_slices((X_train_paths, y_train_genders))
train_ds = train_ds.map(preprocess_image, num_parallel_calls=AUTOTUNE) \
                   .map(lambda x, y: (data_augmentation(x, training=True), y), num_parallel_calls=AUTOTUNE) \
                   .shuffle(buffer_size=1024) \
                   .batch(BATCH_SIZE) \
                   .prefetch(buffer_size=AUTOTUNE)

# 검증 및 테스트 데이터셋 (데이터 증강 미적용)
val_ds = tf.data.Dataset.from_tensor_slices((X_val_paths, y_val_genders))
val_ds = val_ds.map(preprocess_image, num_parallel_calls=AUTOTUNE) \
               .batch(BATCH_SIZE) \
               .prefetch(buffer_size=AUTOTUNE)

test_ds = tf.data.Dataset.from_tensor_slices((X_test_paths, y_test_genders))
test_ds = test_ds.map(preprocess_image, num_parallel_calls=AUTOTUNE) \
                .batch(BATCH_SIZE) \
                .prefetch(buffer_size=AUTOTUNE)

# 7. 딥러닝 모델 구축
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras import regularizers

model = Sequential([
    # 이미지 전처리

    Conv2D(32, (3, 3), activation='relu', input_shape=(IMG_HEIGHT, IMG_WIDTH, CHANNELS)),
    BatchNormalization(), # 배치 정규화 추가
    MaxPooling2D((2, 2)),
    Dropout(0.25),        # 드롭아웃 비율 조정 (선택 사항, 여기서는 Conv2D 이후)

    Conv2D(64, (3, 3), activation='relu'),
    BatchNormalization(), # 배치 정규화 추가
    MaxPooling2D((2, 2)),
    Dropout(0.25),        # 드롭아웃 비율 조정

    Conv2D(128, (3, 3), activation='relu'),
    BatchNormalization(), # 배치 정규화 추가
    MaxPooling2D((2, 2)),
    Dropout(0.25),        # 드롭아웃 비율 조정

    Flatten(),
    Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    BatchNormalization(), # 배치 정규화 추가
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

# 8. 모델 컴파일
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.summary()

# 9. 콜백 설정
checkpoint_filepath = 'best_utkface_cnn_model_augmented.keras'
model_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_filepath,
    save_weights_only=False,
    monitor='val_accuracy',
    mode='max',
    save_best_only=True,
    verbose=1
)

early_stopping_callback = tf.keras.callbacks.EarlyStopping(
    monitor='val_accuracy',
    patience=15, #
    mode='max',
    restore_best_weights=True,
    verbose=1
)

reduce_lr_callback = tf.keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=7,
    min_lr=0.000001,
    verbose=1
)

callbacks_list = [model_checkpoint_callback, early_stopping_callback, reduce_lr_callback]


history = model.fit(
    train_ds,
    epochs=100, # 충분히 큰 에포크 수 (EarlyStopping이 알아서 조기 중단)
    validation_data=val_ds,
    callbacks=callbacks_list,
    verbose=1
)


test_loss, test_accuracy = model.evaluate(test_ds, verbose=0)
print(f"\n테스트 손실: {test_loss:.4f}")
print(f"테스트 정확도: {test_accuracy:.4f}")


plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='훈련 정확도')
plt.plot(history.history['val_accuracy'], label='검증 정확도')
plt.title('CNN 모델 정확도')
plt.xlabel('에포크')
plt.ylabel('정확도')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='훈련 손실')
plt.plot(history.history['val_loss'], label='검증 손실')
plt.title('CNN 모델 손실')
plt.xlabel('에포크')
plt.ylabel('손실')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
